```python
import pandas as pd


insurance = pd.read_csv("types-of-health-insurance-coverage-by-age.csv")


categories_to_plot = [
    'Estimate; Total:',
    'Estimate; Under 18 years:',
    'Estimate; 18 to 34 years:',
    'Estimate; 35 to 64 years:',
    'Estimate; 65 years and over:'
]

sorted_insurance = insurance.sort_values(by='Estimate; Total:', ascending=False)



for category in categories_to_plot:

    top_10 = sorted_insurance.nlargest(10, category)
    

    category_df = pd.DataFrame({
        'Neighborhood': top_10['Neighborhood'],
        category: top_10[category]
    })
    

    print(f"\nTop 10 Neighborhoods by {category}:")
    print(category_df)


```

The estimates of each neighborhood are given above. As seen, Squirrel Hill South has the estimated most total health insurance coverage. However, while that may be, there are specific demographics more at risk to health problems and so the data is also broken down by age group to see which neighborhood has the most health insurance per age group. The danger age groups are typically under 18 and 65 and over, which Squirrel Hill South has the most coverage for both those age groups. 

```python
for category in categories_to_plot:

    top_10 = sorted_insurance.nlargest(10, category)
    

    category_df = pd.DataFrame({
        'Neighborhood': top_10['Neighborhood'],
        category: top_10[category]
    })
    

    category_df.plot(
            x='Neighborhood',
            y=category,
            kind='bar',
            figsize=(16, 8),
            title=f'Top 10 Neighborhoods by {category}',
            legend=False,
            rot=45
        )
```

The graphs just diplay the data in a more presentable fashion then just the raw data. However, some counties may have a larger population than other counties, so I decided to use the health coverage per person, by dividing the total health insurance by the total population. 

```python
pop = pd.read_csv("total-population.csv")

merged_data = pd.merge(insurance, pop[['Neighborhood', 'Estimate; Total']], on='Neighborhood', how='left')

merged_data['Ratio'] = merged_data['Estimate; Total:'] / merged_data['Estimate; Total']

filtered_data = merged_data[merged_data['Ratio'] <= 1]

categories_to_plot = [
    'Estimate; Total:',
]

sorted_filtered_data = filtered_data.sort_values(by='Estimate; Total:', ascending=False)

for category in categories_to_plot:
    top_10 = sorted_filtered_data.nlargest(10, category)
    
    category_df = pd.DataFrame({
        'Neighborhood': top_10['Neighborhood'],
        category: top_10[category],
        'Population Estimate': top_10['Estimate; Total'],
        'Insurance Estimate': top_10['Estimate; Total:'],
        'Ratio': top_10['Ratio']
    })
    
    print(f"\nTop 10 Neighborhoods by {category} with ratio <= 1:")
    print(category_df)

    category_df.plot(
        x='Neighborhood',
        y='Ratio',
        kind='bar',
        figsize=(16, 8),
        title=f'Top 10 Neighborhoods by Ratio for {category} (Ratio < 1)',
        legend=False,
        rot=45
    )
```

As you can see now, the data has changed. Greenfield has the highest ratio, with all of its inhabitants being insured. Despite Squirrel Hill South having the highest total insurance and the most in the dangerous categories in terms of age, Greenfield is actually the safest because in comparison, all of its members, no matter what age, are medically insured. So the data proves that Greenfield is actually the safest neighborhood in terms of medical insurance and safety with an 100% insurance percentage.
