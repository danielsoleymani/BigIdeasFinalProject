The metric that I chose was the number of fires in each neighborhood. The data tracks the number of fire incidents per neighborhood that the Pittsburgh Bureau of Fire responds to and the type of fire. I chose this metric because I would want to know how many fires happen around me before choosing to live in an area. I will be ranking each neighborhood by the number of fires and the types of fires that happen in each neighborhood.

This code collects all of the types of fires and which neighborhood they happen in.

```python
import pandas as pd

%matplotlib inline

fireType=dict()
type = pd.read_csv("Final-Project-data.csv", index_col="type_description")
typeOfFire=[]

for fire, row in type.iterrows():
    typeOfFire.append(fire)

fireType=pd.Series(typeOfFire)
```

This code goes through each fire incident assigning the firetype and the number of each fire to each neighborhood. Building fires and passenger vehicle fires are weighted higher. It then prints out the raw values.

```python

name=dict()

data = pd.read_csv("Final-Project-data.csv", index_col="neighborhood")
count=-1
for index, row in data.iterrows():
    count=count+1
    if index in name:
        if fireType.iloc[count]=="Building fire" or fireType.iloc[count]=="Passenger vehicle fire":
            name[index]==name[index]+5
        else:
            name[index]=name[index]+1
    else:
        if fireType.iloc[count]=="Building fire" or fireType.iloc[count]=="Passenger vehicle fire":
            name[index]=5
        else:
            name[index]=1

for index in name:
    print(index,":",name[index])


```

This code sorts all the neighborhood by least to most fire incidents

```python
values=dict(sorted(name.items(), key = lambda item: item[1]))

count=1

for index in values:
    values[index]=count
    print(index,":",values[index])
    count=count+1
```

In conclusion using the metric of fire incidents per neighborhood and weighing the incidents based on the severity the best neighborhood is Mt.Oliver due to having the lowest score.
